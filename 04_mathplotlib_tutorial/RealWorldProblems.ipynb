# Project 2: Fifa

# Histogram

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall,bins=bins,color="r")

plt.xticks(bins)

plt.xlabel("Skill Level")
plt.ylabel("Number of Players")
plt.title("Distribution of Player Skills in FIFA 2018")

plt.show()
```

# Pie chart

```python
fifa["Preferred Foot"]
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right= fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

# print(left)
# print(right)

labels= ['Left','Right']
color = ['r','g']

plt.title("Foot Preference of Players in FIFA")

plt.pie([left,right],labels=labels,colors = color,autopct='%1.2f%%',shadow= True, explode=(0,0.1))
plt.show()
```

# Pie chart 2

```python
print(fifa.Weight) 
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight] # weight was not a str type
fifa.Weight
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.figure(figsize=(9,6))

plt.style.use('ggplot')

light = fifa.loc[(fifa.Weight < 125)].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150 )].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175 )].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200 )].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['under 125','125-150','150-175','175-200','over 200']
# color = ['red','blue','green','yellow','lightblue']

plt.title("Weight Distribution of FIFA Players ")

plt.pie(weights,labels=labels,shadow = True,autopct='%1.2f%%',pctdistance=0.8,explode=(0.3,0.2,0.2,0,0.3))
plt.show()
```
