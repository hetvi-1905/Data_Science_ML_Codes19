# Formating and styling a simple Graph with two lines

```python
from matplotlib import pyplot as plt 
from matplotlib import style

style.use("ggplot")

x=[5,8,10]
y=[12,16,6]

x2=[6,9,11] 
y2=[16,15,7]

plt.plot(x,y,label="line one", linewidth=5) 
plt.plot(x2,y2,label="line two",linewidth=5)

plt.title("Epic Info")

plt.ylabel("Y axis") 
plt.xlabel("X axis")

plt.legend()

plt.grid(True,color="k")

plt.show()
```

# Histogram

# Stack plot

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([],[],color='m', label="Sleeping", linewidth=5)  # all these are small graphs of width = 5 used as legend on the corner
plt.plot([],[],color='c', label="Eating", linewidth=5) 
plt.plot([],[],color='r', label="Working", linewidth=5) 
plt.plot([],[],color='k', label="Playing", linewidth=5)

plt.stackplot(days, sleeping ,eating ,working ,playing ,colors=['m','c','r','k'])

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.title('stack Plot')

plt.legend()

plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt

slices=[25,25,25,25]

activities=['sleeping', 'eating', 'working','playing']
cols=['c','m','r','b']

plt.pie(slices,
    labels=activities,
    colors=cols,
    startangle=90,
    shadow= True,
    explode=(0,0.1,0,0), # highlighted the 2nd slice i.e eating as it is 0.1 and others are 0
    autopct='%1.1f%%')

plt.title('Pie Plot') 
plt.show()
```

# Multiple plots together

```python
import numpy as np 
import matplotlib.pyplot as plt

def f(t):
    return t**(1/2)  # square-root

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)  # -------------> see this in more detail to understand more....
plt.plot(t1, f(t1), 'ro')  # red + linestyle"o"

plt.subplot(212) 
plt.plot(t2, t2**2,'b--') # blue + linestyle"--"

#  subplot = 221 and 222 together gives different output and 211 and 212 together gives different o/p

plt.show()
```

# Basic Graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y,label='2x',color="red", linewidth=2, marker=".",linestyle="--",markersize="10",markeredgecolor="blue")

plt.title("Our first graph!",fontdict={'fontname':'Comic Sans MS','fontsize':30,'fontweight':100,'color':'red','verticalalignment': 'baseline'})

plt.xlabel("My x-axis",fontdict={'fontname':'Arial','fontsize':10})
plt.ylabel("My y-axis",fontdict={'fontname':'Arial','fontsize':10})

plt.xticks([0,1,2,3,7,8])  # marks on x-axis
plt.yticks([0,2,4,6,7,7.5,8])  # marks on y-axis

plt.legend()

plt.show()
```

```python
#resize your graph
plt.figure(figsize=(5,3),dpi=200)  # dpi = pixels per inch

plt.title("Our first graph!",fontdict={'fontname':'Comic Sans MS','fontsize':15,'fontweight':100,'color':'red','verticalalignment': 'baseline'})

# line Number 1
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y,label='2x',color="red", linewidth=2, marker=".",linestyle="--",markersize="10",markeredgecolor="blue")

# Line number 2
x2 = np.arange(0,4,0.5)  # 0 to 4 with gaps og 0.5 between each element 0,0.5,1,1.5,2,2.5,3,3.5,4 -->9 points

# x2= [0,0.5,1,1.5,2,2.5,3,3.5,4] and y2 = [x2**2]
plt.plot(x2[0:5],x2[0:5]**2,label="2*x",color="blue",linewidth=2,marker=".",markersize="10",markeredgecolor="red") # half with different style
plt.plot(x2[4:],x2[4:]**2,color="blue",linewidth=2,marker=".",linestyle="--",markersize="10",markeredgecolor="red") #other half with different style

                       
plt.xlabel("My x-axis",fontdict={'fontname':'Arial','fontsize':10})
plt.ylabel("My y-axis",fontdict={'fontname':'Arial','fontsize':10})

# plt.xticks([0,1,2,3,7,8])  # marks on x-axis
plt.yticks([0,2,4,6,7,7.5,8])  # marks on y-axis

plt.legend()

plt.savefig("mygraph.png",dpi=200)  # to save this graph pic in png format in the current directory that we are working inS

plt.show()
```

# Bar Chart/Graph

```python
labels = ['A','B','C']
values = [2,5,4]

bars = plt.bar(labels,values)

# way 1
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*') # hatches are = - , / , . , | ,etc...

# way 2
patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0)) # pop(0) = line wise patterns , pop(-1) = reverse the order of patterns ,pop() = random patterns order

plt.figure(figsize=(6,4),dpi=100)

plt.show()
```
