```python
# joint plot
sns.jointplot(x="speeding",y="alcohol",data=crash_df,kind="kde")
sns.jointplot(x="speeding",y="alcohol",data=crash_df,kind="reg")
sns.jointplot(x="speeding",y="alcohol",data=crash_df,kind="hex")
sns.jointplot(x="speeding",y="alcohol",data=crash_df,kind="hist")
```

```python
# pair plot
sns.pairplot(crash_df)
```

```python
# pairplot 2
tips_df = sns.load_dataset('tips')
print(tips_df)

sns.pairplot(data=tips_df,hue='sex',palette="Blues")
```

```python
# rug plot(2 syntaxs)
sns.rugplot(tips_df['tip'])
plt.show()
sns.rugplot(data=tips_df,x='tip')
plt.show()
```

```python
# styling
sns.set_style("ticks")

plt.figure(figsize=(10,5))

sns.set_context("talk",font_scale=1.5)

sns.jointplot(x='speeding',y="alcohol",data=crash_df,kind="reg")

sns.despine(left=True,bottom=True)
```

```python
# categorical plots

#bar plots

sns.barplot(x="sex",y="total_bill",data=tips_df,estimator=np.cov)
```

```python
# count plot
sns.countplot(x='sex',data=tips_df)
```

```python
# box plot
sns.boxplot(x="day",y="total_bill",data=tips_df,hue='sex')

plt.legend(loc=0)
```

```python
# violin plot
sns.violinplot(x="day",y="total_bill",data=tips_df,hue="sex",split=True)
```

```python
# strip plot
plt.figure(figsize=(8,5))

sns.stripplot(x="day",y="total_bill",data=tips_df,jitter=True,hue="sex",dodge=True)
```

```python
# swarm plot

sns.swarmplot(x="day",y="total_bill",data=tips_df)
```

```python
# palettes

plt.figure(figsize=(8,5))
sns.set_style("dark")
sns.set_context("talk")
sns.stripplot(x="day",y="total_bill",data=tips_df,hue="sex",palette="magma")
plt.legend(loc=1)
```

```python
# matrix plot
# heatmaps
plt.figure(figsize=(8,6))
sns.set_context("paper",font_scale=1.4)

crash_mx = crash_df.corr()
# crash_mx

sns.heatmap(crash_mx,annot=True,cmap="Blues")
```

```python
#heat map 2
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index="month",columns="year",values="passengers")
sns.heatmap(flights,cmap="Blues",linecolor="white",linewidth=1)
```

```python
#cluster map

iris = sns.load_dataset("iris")

species = iris.pop("species")
sns.clustermap(iris)
```

```python
# pairgrids ### IMPORTANT ###
iris = sns.load_dataset("iris")
iris_g = sns.PairGrid(iris,hue="species")    # all different subtypes important to know , run them individualy to know each
# iris_g.map(plt.scatter)
# iris_g.map_diag(plt.hist)
# iris_g.map_offdiag(plt.scatter)   
# iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
iris_g = sns.PairGrid(iris,hue="species",x_vars=["sepal_length","sepal_width"],y_vars=["petal_length","petal_width"])
iris_g.map(plt.scatter)
iris_g.add_legend()
```
