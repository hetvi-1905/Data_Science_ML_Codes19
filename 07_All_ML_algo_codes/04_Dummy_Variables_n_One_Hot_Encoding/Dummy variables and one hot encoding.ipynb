```python
import pandas as pd
df = pd.read_csv("homeprices_2.csv")
df.head()
# df.columns
```

```python
dummies = pd.get_dummies(df.town)  # copy paste the names of column as there can be some space mistake and you will not get o/p
dummies
```

```python
merged_df = pd.concat([df,dummies],axis="columns")
merged_df
```

```python
final_df = merged_df.drop(columns=['town',"calgary"],axis="columns")  # dummy variable trap avoiding
final_df
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final_df.drop("price",axis="columns")
X
```

```python
y = final_df.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2000,0,1]])  # cost of house in toronto
```

```python
model.predict([[2000,0,0]])   # cost of house in calgary
```

```python
model.predict([[2000,1,0]])   # cost of house in ontario
```

```python
model.score(X,y)    # accuracy 
```
