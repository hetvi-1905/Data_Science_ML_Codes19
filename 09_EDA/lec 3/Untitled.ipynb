# Feature Engineering process|

## 1)Date_of_journey

```python
final_df["Date_of_Journey"].str.split("/")
```

```python
final_df["Date_of_Journey"].str.split("/").str[0]
```

```python
final_df["Date_of_Journey"].str.split("/").str[1]
```

```python
final_df["Date"]= final_df["Date_of_Journey"].str.split("/").str[0]
final_df["Month"]= final_df["Date_of_Journey"].str.split("/").str[1]
final_df["Year"]= final_df["Date_of_Journey"].str.split("/").str[2]
```

```python
final_df.head(3)
```

```python
# # method 2 using lambda func...
# final_df["Date"] = final_df["Date_of_Journey"].apply(lambda x : x.split("/")[0])
# final_df["Month"] = final_df["Date_of_Journey"].apply(lambda x : x.split("/")[1])
# final_df["Year"] = final_df["Date_of_Journey"].apply(lambda x : x.split("/")[2])
# final_df.head(3)
```

```python
final_df["Date"]=final_df["Date"].astype(int)
final_df["Month"]=final_df["Month"].astype(int)
final_df["Year"]=final_df["Year"].astype(int)
final_df.info()
```

```python
final_df.drop(["Date_of_Journey"],axis="columns",inplace=True)
final_df.head(3)
```

```python
final_df["Arrival_hour"]=final_df["Arrival_hour"].astype(int)
final_df["Arrival_minutes"]=final_df["Arrival_minutes"].astype(int)
final_df.info()
```

# 3)Departure_time

```python
final_df["Dep_Time"] = final_df["Dep_Time"].apply(lambda x:x.split(" ")[0])

final_df["Dep_hour"] = final_df["Dep_Time"].apply(lambda x:x.split(":")[0])
final_df["Dep_minutes"] = final_df["Dep_Time"].apply(lambda x:x.split(":")[1])

final_df["Dep_hour"]=final_df["Dep_hour"].astype(int)
final_df["Dep_minutes"]=final_df["Dep_minutes"].astype(int)
final_df.info()
```

```python
final_df.drop(["Dep_Time"],axis="columns",inplace=True)
final_df.head(3)
```
